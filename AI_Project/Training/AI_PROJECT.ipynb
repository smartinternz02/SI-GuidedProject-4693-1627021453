# Performing Train and Test Split


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
x_test
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_test
```

```python
x_train
```

# BUILDING ANN MODEL


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 10, kernel_initializer = "random_uniform",activation ="relu"))
```

```python
model.add(Dense(units = 30 , kernel_initializer = "random_uniform",activation ="relu"))
```

```python
model.add(Dense(units = 20 , kernel_initializer = "random_uniform",activation ="relu"))
```

```python
model.add(Dense(units=1, kernel_initializer="random_uniform", activation="sigmoid"))
```

```python
model.compile(optimizer = "rmsprop",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(x_train,y_train, batch_size =32,epochs = 200, validation_data = (x_test,y_test))
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred>0.5
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label = "train")
plt.plot(history.history['val_accuracy'],label="test")
plt.legend()
plt.show()
```
